```python
import pandas as pd
import hvplot.pandas 
df = pd.read_csv("https://mesonet.agron.iastate.edu/cgi-bin/request/asos.py?station=ORD&data=all&year1=2017&month1=8&day1=10&year2=2017&month2=8&day2=13&tz=Etc%2FUTC&format=onlycomma&latlon=no&elev=no&missing=M&trace=T&direct=no&report_type=3&report_type=4")
```

```python
df_ord = df.loc[df.station == "ORD"]
```

```python
df_ord["time"] = pd.to_datetime(df_ord.valid)
df_ord = df_ord.set_index("time")
```

```python
temp = df_ord.tmpf.hvplot()
speed = df_ord.sknt.hvplot()
cover = df_ord.skyc1.hvplot()

(temp+speed+cover).cols(1)
```
